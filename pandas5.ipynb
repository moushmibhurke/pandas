```python
df= pd.read_csv("C:\pythoncode\data\weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'no event'
    })
new_df #passing dictionary 
```

```python
new_df = df.fillna(method="ffill") #forwardfill
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") #backwardfill
new_df
```

```python
new_df = df.fillna(method="ffill", limit=1) #forward limit
new_df
```

```python
new_df =df.interpolate() #intermidiate data point
new_df
```

```python
new_df =df.interpolate(method = 'time') #intermidiate data point
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how ="all") 
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
