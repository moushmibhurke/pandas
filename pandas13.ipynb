```python
import pandas as pd
df = pd.read_excel("C:\pythoncode\data\survey.xlsx")
df
```

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

```python
pd.crosstab(df.Sex,[df.Nationality,df.Handedness]) #column level array
```

```python
pd.crosstab([df.Nationality,df.Sex],df.Handedness) #row lelvel array
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index') #normalize shift+tab
```

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
